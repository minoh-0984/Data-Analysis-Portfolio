## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
QUERY = """
SELECT * FROM master
"""

df = pd.read_sql_query(QUERY,conn)

df.hist(column='ConvertedComp')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age FROM master
"""

df = pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(10,5))
sns.boxplot(x=df.Age, data=df)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """

SELECT Age,ConvertedComp,WorkWeekHrs,CodeRevHrs,MainBranch

FROM master

"""

df=pd.read_sql_query(QUERY,conn)

df

df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='darkblue')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """

SELECT Age,ConvertedComp,WorkWeekHrs,CodeRevHrs

FROM master

"""

df = pd.read_sql_query(QUERY,conn)

x='WorkWeekHrs'
y='CodeRevHrs'
s='Age'

bubble=plt.scatter(x,y,s,alpha=0.5, data=df)
plt.xlabel("CodeRevHrs", size=16)
plt.ylabel("WorkWeekHrs", size=16)
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import numpy as np
QUERY = """

select count(DatabaseDesireNextYear) as Count,DatabaseDesireNextYear
from DatabaseDesireNextYear 
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) DESC LIMIT 5

"""

df2=pd.read_sql_query(QUERY,conn)

df2.set_index('DatabaseDesireNextYear',inplace=True)

df2

y = np.array([4328, 3649, 3331, 3281, 2856])
mylabels = ["PostgreSQL", "MongoDB", "Redis", "MySQL", "Elaticsearch"]

plt.pie(y, labels=mylabels, startangle= 90)
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master 
WHERE Age between 30 and 35
"""

df_stacked = pd.read_sql_query(QUERY,conn)

df_z=df_stacked[(df_stacked["Age"]>29) & (df_stacked["Age"]<36)]

grouped=df_z.groupby(by='Age')
groupedm=grouped.median()

groupedm.plot(kind='bar', stacked=True)
plt.title("Median Work Week Hrs & CodeRevHrs by Age")
plt.xlabel("Age")
plt.ylabel("Hrs")
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master 
WHERE Age between 45 and 60
"""

df_stacked = pd.read_sql_query(QUERY,conn)

df_z=df_stacked[(df_stacked["Age"]>=45) & (df_stacked["Age"]<=60)]

grouped=df_z.groupby(by='Age')
groupedm=grouped.median()

groupedm.plot(kind='line', stacked=True)
plt.title("Median ConvertedComp for age 45~60")
plt.xlabel("Age")
plt.ylabel("ConvertedComp")
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
table_name = 'master'

QUERY="""
SELECT MainBranch,count(MainBranch)
FROM master
group by MainBranch
"""

df=pd.read_sql_query(QUERY,conn)
df.head(5)
df.plot(kind='barh', figsize=(12,12))
plt.xlabel('MainrBranch')
plt.show()
```
